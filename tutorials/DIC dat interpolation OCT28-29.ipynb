# MK-Xray tester and strain interpolation
- An issue with NIST MK-Xray tester is that the central area of the specimen doesn't provide the strain because speckled patterns are removed thereon to allow the X-ray beam.
- In the following, an example to interpolate the missing strain data

## 1. We assume that the data file is a collection of 'node' data resulting from various DIC images.

```python
import scipy; from scipy import interpolate
import MKDIC
from MKDIC import readers

# read data file
dat=readers.read_disc(fn='../dat/IFsteel/2014OCT28_IFsteel/DiscNodeDat_2014OCT28.csv',ndat_col=6)
```

## 2. The data has got ndat_col consisting of , X, Y, exx, eyy, exy, flag.

```python
x=dat[:,:,0]
y=dat[:,:,1]
exx=dat[:,:,2]
eyy=dat[:,:,3]
exy=dat[:,:,4]
ezz=-exx-eyy
```

## 3. Load up the data and plot DIC results

```python
## Define DIC_plot
def dic_plot(ax,X,Y,Z,lab=None):
    mn = min(Z);mx = max(Z)
    norm = mpl.colors.Normalize(vmin=mn,vmax=mx)
    cmap, m = mpl_lib.norm_cmap(mn=mn,mx=mx,cm_name='brg')

    for i in range(len(X)):
        try:
            c=m.to_rgba(Z[i])
            ax.plot(X[i],Y[i],'o',ms=5,mec='None',color=c)
        except:pass
    #ax.set_aspect('equal')
    ax.set_xlabel('x coordinate [mm]')
    ax.set_ylabel('Y coordinate [mm]')
    b = ax.get_position()
    w0 = b.x1+0.002
    h0 = b.y0
    h = b.y1-b.y0

    axcb = plt.gcf().add_axes([w0,h0,0.015,h])
    mpl_lib.add_cb(axcb,cmap=cmap,filled=True,
                   norm=norm,ylab=lab,format='%5.3f')

import MP; from MP.lib import mpl_lib # personal library to deco
import matplotlib.gridspec as gridspec

istp=-1 ## The last image data


## Plot experimental data
# - prepare grids subplots
gs = gridspec.GridSpec(1,4)
gs.update(left=0.05,hspace=0.20,wspace=0.7)
plt.figure(figsize=(20,3))
ax1=plt.subplot(gs[0]);ax2=plt.subplot(gs[1]);
ax3=plt.subplot(gs[2]);ax4=plt.subplot(gs[3]);
# - run dic_plot for exx,eyy,ezz,exy
dic_plot(ax1,x[istp],y[istp],ezz[istp])#,lab=r'$\bar{E}_{zz}$')
dic_plot(ax2,x[istp],y[istp],exx[istp])#,lab=r'$\bar{E}_{xx}$')
dic_plot(ax3,x[istp],y[istp],eyy[istp])#,lab=r'$\bar{E}_{yy}$')
dic_plot(ax4,x[istp],y[istp],exy[istp])#,lab=r'$\bar{E}_{xy}$')

ax1.set_title(r'$\bar{E}_{zz}$',fontsize=12)
ax2.set_title(r'$\bar{E}_{xx}$',fontsize=12)
ax3.set_title(r'$\bar{E}_{yy}$',fontsize=12)
ax3.set_title(r'$\bar{E}_{xy}$',fontsize=12)
```

```python
from scipy.interpolate import Rbf
istp = -1

X=x[istp][::]
Y=y[istp][::]
Ezz=ezz[istp][::]
Exx=exx[istp][::]
Eyy=eyy[istp][::]
Exy=exy[istp][::]

Xcord, Ycord, e33, e11, e12, e22 = [], [], [], [], [], []
for i in range(len(X)):
    if not(np.isnan(X[i]) or np.isnan(Y[i]) or np.isnan(Ezz[i])):
        Xcord.append(X[i])
        Ycord.append(Y[i])
        e33.append(Ezz[i])
        e11.append(Exx[i])
        e22.append(Eyy[i])
        e12.append(Exy[i])

Xcord=np.array(Xcord)
Ycord=np.array(Ycord)
E11=np.array(e11)
E22=np.array(e22)
E12=np.array(e12)
E33=np.array(e33)
```

```python
def Rbf_intp(x,y,z,l=-13,u=13,n=30):
    """
    Apply Radial basis function interpolation 
    on the 2D scattered data such as z=z(x,y)
    
    arguments
    =========
    x,y,z
    l=-13
    u=13
    n=30
    
    l and u defines the upper and lower rectangular
    boundary in (x,y) plane, in which the interpolation
    is conducted: l<x<u; l<y<u.
    n defines the meshgrid size: the bigger the n value,
    the finer the grid.
    
    returns the interpolated z(x,y) in the n x n grid.
    """
    import scipy; from scipy import interpolate
    from scipy.interpolate import Rbf
    tck = interpolate.bisplrep(x=x,y=y,z=z)
    ti = np.linspace(l,u,n)
    xi,yi = np.meshgrid(ti,ti)
    
    rbf = Rbf(x,y,z,epsilon=1)
    zi = rbf(xi,yi)
    return xi,yi,zi

def Rbf_plot(ax,x,y,z,iopt=0,lab=None,norm=None,cmap=None,m=None,
             icb=True,irm=True):
    from MP.lib import mpl_lib
    try:
        mn = min(z);mx = max(z)
    except:
        mn = min(z.flatten())
        mx = max(z.flatten())
    if type(norm)==type(None):
        norm = mpl.colors.Normalize(vmin=mn,vmax=mx)
    
    CM, M = mpl_lib.norm_cmap(mn=mn,mx=mx,cm_name='brg')
    if type(cmap)==type(None):
        cmap = CM
    if type(m)==type(None):
        m = M
    
    if iopt==0: ax.pcolor(x,y,z,cmap=cmap)    
    if iopt==1:
        for i in range(len(x)):
            try:
                c=m.to_rgba(z[i])
                ax.plot(x[i],y[i],'o',mec='None',ms=3,color=c);
            except: pass
    
    #ax.set_aspect('equal');
    if not(irm):
        ax.set_title(lab)

    #ax.set_xlabel('x coordinate [mm]'); ax.set_ylabel('Y coordinate [mm]')
    
    if icb:
        b = ax.get_position()
        w0 = b.x1+0.01
        h0 = b.y0
        h  = b.y1-b.y0
        dim = [w0,h0,0.02,h]        
        axcb = plt.gcf().add_axes(dim)
        mpl_lib.add_cb(axcb,cmap=cmap,filled=True,
                       norm=norm,ylab=lab,format='%5.3f')
        
    return norm, cmap, m

    
## Application of the above two definitions
gs = gridspec.GridSpec(3,4)
gs.update(left=0.05,right=0.90,hspace=0.0,wspace=0.15)
fig=plt.figure(figsize=(6,3))
ax1=fig.add_subplot(gs[0]);ax2=fig.add_subplot(gs[1]);
ax3=fig.add_subplot(gs[2]);ax4=fig.add_subplot(gs[3]);
ax5=fig.add_subplot(gs[4]);ax6=fig.add_subplot(gs[5]);
ax7=fig.add_subplot(gs[6]);ax8=fig.add_subplot(gs[7]);
ax9=fig.add_subplot(gs[8]);ax10=fig.add_subplot(gs[9]);
ax11=fig.add_subplot(gs[10]);ax12=fig.add_subplot(gs[11]);
#ax13=fig.add_subplot(gs[12]);ax14=fig.add_subplot(gs[13]);
#ax15=fig.add_subplot(gs[14]);
axes = fig.axes

l=-13; u=13;n = 30

xi,yi,E11_i = Rbf_intp(Xcord,Ycord,E11,l=l,u=u,n=n)
xi,yi,E22_i = Rbf_intp(Xcord,Ycord,E22,l=l,u=u,n=n)
xi,yi,E33_i = Rbf_intp(Xcord,Ycord,E33,l=l,u=u,n=n)
xi,yi,E12_i = Rbf_intp(Xcord,Ycord,E12,l=l,u=u,n=n)


norm, cmap, m = Rbf_plot(ax1,Xcord,Ycord,E11,1,lab=r'$\bar{E}_{11}$')
Rbf_plot(ax5,xi,yi,E11_i,0,lab=r'$\bar{E}_{11}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax9,Xcord,Ycord,E11,1,lab=r'$\bar{E}_{11}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax9,xi,yi,E11_i,0,lab=r'$\bar{E}_{11}$',norm=norm,cmap=cmap,m=m,icb=True)

norm, cmap, m = Rbf_plot(ax2,Xcord,Ycord,E22,1,lab=r'$\bar{E}_{22}$')
Rbf_plot(ax6,xi,yi,E22_i,0,lab=r'$\bar{E}_{22}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax10,Xcord,Ycord,E22,1,lab=r'$\bar{E}_{22}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax10,xi,yi,E22_i,0,lab=r'$\bar{E}_{22}$',norm=norm,cmap=cmap,m=m,icb=True)

norm, cmap, m = Rbf_plot(ax3,Xcord,Ycord,E12,1,lab=r'$\bar{E}_{12}$')
Rbf_plot(ax7,xi,yi,E12_i,0,lab=r'$\bar{E}_{12}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax11,Xcord,Ycord,E12,1,lab=r'$\bar{E}_{12}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax11,xi,yi,E12_i,0,lab=r'$\bar{E}_{12}$',norm=norm,cmap=cmap,m=m,icb=True)

norm, cmap,m=Rbf_plot(ax4,Xcord,Ycord,E33,1,lab=r'$\bar{E}_{33}$')
Rbf_plot(ax8,xi,yi,E33_i,0,lab=r'$\bar{E}_{33}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax12,Xcord,Ycord,E33,1,lab=r'$\bar{E}_{33}$',norm=norm,cmap=cmap,m=m)
Rbf_plot(ax12,xi,yi,E33_i,0,lab=r'$\bar{E}_{33}$',norm=norm,cmap=cmap,m=m,icb=True)

mpl_lib.rm_all_lab(fig.axes)
ft=dict(fontsize=5)
ax9.set_xlabel('X coordinate [mm]',ft); ax9.set_ylabel('Y coordinate [mm]',ft)

from MP.lib import mpl_lib
mpl_lib.tune_xy_lim(axes)
```

```python
def add_cb(ax,cmap,norm,ylab=r'$\Delta\bar{E}_{33}$',fmt='%+8.2e'):
    b = ax.get_position()
    w0 = b.x1+0.002
    h0 = b.y0
    h  = b.y1-b.y0    
    dim = [w0,h0,0.008,h]        
    axcb = plt.gcf().add_axes(dim)
    mpl_lib.add_cb(axcb,cmap=cmap,filled=True,
                   norm=norm,ylab=ylab,format=fmt)
```

```python
## example to compare intp(-e11-e22) and with (-intp(e11)-intp(e22))
gs = gridspec.GridSpec(1,5)
gs.update(left=0.05,hspace=0.20,wspace=0.8)
fig = plt.figure(3,figsize=(20,2.5))
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])
ax3 = fig.add_subplot(gs[2])
ax4 = fig.add_subplot(gs[3])
ax5 = fig.add_subplot(gs[4])
axes= fig.axes

norm,cmap,m = Rbf_plot(ax2,xi,yi,E11_i,0,lab=r'intp$(\bar{E}_{11})$')
add_cb(ax2,cmap=cmap,norm=norm,ylab=r'$\bar{E}_{11}$',fmt='%5.3f')
norm,cmap,m = Rbf_plot(ax3,xi,yi,E22_i,0,lab=r'intp$(\bar{E}_{22})$')
add_cb(ax3,cmap=cmap,norm=norm,ylab=r'$\bar{E}_{22}$',fmt='%5.3f')
norm,cmap,m = Rbf_plot(ax1,xi,yi,E33_i,0,lab=r'intp$(\bar{E}_{33})$')
add_cb(ax1,cmap=cmap,norm=norm,ylab=r'$\bar{E}_{33}$',fmt='%5.3f')


intpd=-E11_i-E22_i
ax4.pcolor(xi,yi,intpd,cmap=cmap)
ax4.set_title(r'-intp$(\bar{E}_{11})-\mathrm{intp}(\bar{E}_{22})$')
add_cb(ax4,cmap=cmap,norm=norm,ylab=r'$\bar{E}_{33}$',fmt='%5.3f')
ax4.set_xlabel('x coordinate [mm]'); ax4.set_ylabel('Y coordinate [mm]')
 
err = E33_i - intpd
norm,cmap,m = Rbf_plot(ax5,xi,yi,err,0,lab=r'$\Delta\bar{E}_{33}=$') #-intp$(\bar{E}_{11})-\mathrm{intp}(\bar{E}_{22})-$intp$(\bar{E}_{33})$')
add_cb(ax5,cmap=cmap,norm=norm,ylab=r'$\Delta\bar{E}_{33}$',fmt='%+8.2e')

```

```python
def avg(r=5,ax=None,iplot=True,x=None,y=None,z=None,
        n=20,l=-8,u=8,x0=0,y0=0,lab='None'):
    mn = min(z);mx = max(z)
    if iplot:
        norm = mpl.colors.Normalize(vmin=mn,vmax=mx)
        cmap, m = mpl_lib.norm_cmap(mn=mn,mx=mx,cm_name='brg')
        Rbf_plot(ax,x,y,z,1,lab=lab,norm=norm,cmap=cmap,m=m)

    xi, yi, zi = Rbf_intp(x,y,z,l=l,u=u,n=n)

    dat=[]
    for i in range(len(xi)):
        for j in range(len(xi[i])):
            X=xi[i,j]; Y=yi[i,j]
            R=sqrt((X-x0)**2+(Y-y0)**2)
            Z=zi[i,j]
            if iplot: c=m.to_rgba(Z)
            if R<r and iplot:
                ax.plot(X,Y,'.',ms=5,mfc=c,mec='None')
            if R<r:
                dat.append(Z)

    return np.array(dat).mean()

gs = gridspec.GridSpec(1,3)
gs.update(left=0.05,hspace=0.10,wspace=0.3)
plt.figure(figsize=(20,5))
ax1=plt.subplot(gs[0]);ax2=plt.subplot(gs[1]);
ax3=plt.subplot(gs[2])

avg(r=5,ax=ax1,iplot=True,x=Xcord,y=Ycord,z=E11,n=50,l=-13,u=13,x0=1,y0=0,lab=r'$\bar{E}_{11}$')
ax1.set_aspect('equal')
avg(r=5,ax=ax2,iplot=True,x=Xcord,y=Ycord,z=E22,n=50,l=-13,u=13,x0=1,y0=0,lab=r'$\bar{E}_{11}$')
ax2.set_aspect('equal')
avg(r=5,ax=ax3,iplot=True,x=Xcord,y=Ycord,z=E33,n=50,l=-13,u=13,x0=1,y0=0,lab=r'$\bar{E}_{11}$')
ax3.set_aspect('equal')
```
